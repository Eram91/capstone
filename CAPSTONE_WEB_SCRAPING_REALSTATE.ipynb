```python
# importing libraries for web scraping
import selenium 
import pandas as pd 
import time 
from bs4 import BeautifulSoup 

# importing selenium webdriver 
from selenium import webdriver

# importing required Exceptions which needs to handled 
from selenium.common.exceptions import NoSuchElementException

# importing request 
import requests 
from selenium.webdriver.common.by import By 

import warnings
warnings.filterwarnings('ignore')
```

***Opening Web page on automated chrome browser***

```python
driver = webdriver.Chrome()
driver.get("https://www.realestateindia.com/")
```

***Scraping data for 1st city - Delhi***

```python
#selecting location tab in menubar and clicking on it

search=driver.find_element(By.XPATH, '/html/body/header/div/div/div[2]/div[2]/ul/li[2]/span')
search.click()
```

```python
#selecting 'properties for rent in delhi' from the rent tab and clicking on it

search1=driver.find_element(By.XPATH, '/html/body/header/div/div/div[2]/div[2]/ul/li[2]/div/div/div/ul/li[1]/a')
search1.click()
```

```python
#selecting 'flats for rent in delhi' tab

search2=driver.find_element(By.XPATH, '/html/body/form/div[1]/div/div/ul/li[1]/a/span')
search2.click()
```

```python
#clicking on it 

search3=driver.find_element(By.XPATH, '/html/body/form/div[1]/div/div/ul/li[1]/div/div[1]/ul[1]/li[2]/div/label')
search3.click()
```

```python
#scraping all the urls having rentals properties

product_url=[]
start=0
end=15
for page in range(start,end):
    tags_urls=driver.find_elements(By.XPATH,'//div[@class="prop-clasified ps-list"]')
    for i in tags_urls:
        product_url.append(i.get_attribute('data-url'))
    next_button=driver.find_elements(By.XPATH,'/html/body/section[1]/div/div[2]/div[1]/div[27]/a')
```

```python
#checking the no of urls fetched

len(product_url)
```

```python
# opening first page to scrap required features.

rent_page=driver.find_element(By.XPATH,'/html/body/section[1]/div/div[2]/div[1]/div[4]/div[1]')
rent_page.click()
```

```python
#fetching the data for required features

address=[]
city=[]
rental_price=[]
deposit=[]
no_of_beds=[]
no_of_baths=[]
area=[]
parking=[]
storage=[]
fireplace=[]

for url in product_url:
    driver.get(url)
    time.sleep(2)
    
    try:
        addresses=driver.find_element(By.XPATH,'//ul[@class="pdoi-list"]/li[1]/div')  #for address
        address.append(addresses.text)
    except NoSuchElementException:
        address.append('-')
        
    try:
        cities=driver.find_element(By.XPATH,'/html/body/header/div/div/div[2]/div[1]/a/span') #for city
        city.append(cities.text)
    except NoSuchElementException:
        city.append('-')
        
    try:
        rental=driver.find_element(By.XPATH,'//div[@class="pdt-price"]') #for rental price
        rental_price.append(rental.text)
    except NoSuchElementException:
        rental_price.append('-')
        
    try:
        deposits=driver.find_element(By.XPATH,'//span[@class="app-val"]') #for booking ie deposit
        deposit.append(deposits.text)
    except NoSuchElementException:
        deposit.append('-')
        
    try:
        bed=driver.find_element(By.XPATH,'//ul[@class="pdsi-list"]/li[1]/strong') #for no of beds
        no_of_beds.append(bed.text)
    except NoSuchElementException:
        no_of_beds.append('-')
        
    try:
        bath=driver.find_element(By.XPATH,'//ul[@class="pdsi-list"]/li[2]/strong') #for no of bath
        no_of_baths.append(bath.text)
    except NoSuchElementException:
        no_of_baths.append('-')
        
    try:
        areas=driver.find_element(By.XPATH,'//ul[@class="pdsi-list"]/li[4]/strong') #for area
        area.append(areas.text)
    except NoSuchElementException:
        area.append('-')
        
    try:
        parkings=driver.find_element(By.XPATH,'//ul[@class="pd-aminities-list"]/li/ul/li[1]') #for parking
        parking.append(parkings.text)
    except NoSuchElementException:
        parking.append('-')
        
    try:
        storages=driver.find_element(By.XPATH,'//ul[@class="ap-list"]/li[3]/span[2]') #for storage
        storage.append(storages.text)
    except NoSuchElementException:
        storage.append('-')
        
    try:
        fireplaces=driver.find_element(By.XPATH,'//ul[@class="pd-aminities-list"]/li/ul/li[9]') #for fireplace
        fireplace.append(fireplaces.text)
    except NoSuchElementException:
        fireplace.append('-')
```

```python
#Creating data frame for table1

d1= pd.DataFrame({'Address':address,'City':city,'Rental_Price':rental_price,'Deposit':deposit})
```

```python
#adding columns of state_code and country 

delhi1 = d1.reindex(d1.columns.tolist() + ["State_code", "Country"], axis=1)
```

```python
delhi1['Country']="India"
```

```python
delhi1['State_code']="7"
```

```python
delhi1
```

```python
#creating data frame for table2

d2= pd.DataFrame({'No_of_beds':no_of_beds,'No_of_baths':no_of_baths,'Area':area})
```

```python
#adding column 'pets_allowed'

delhi2 = d2.reindex(d2.columns.tolist() + ["Pets_allowed"], axis=1)
```

```python
delhi2
```

```python
#creating data frame for table3

d3= pd.DataFrame({'Parking':parking,'Storage':storage,'Fireplace':fireplace})
```

```python
#adding column 'washer/dryer, ac, dishwasher, hardwoodfloors, roofdeck'

delhi3 = d3.reindex(d3.columns.tolist() + ["Washer/Dryer","AC","Dishwasher","Hardwoodfloors","Roofdeck"],axis=1)
```

```python
delhi3
```

```python
delhi1.to_csv('delhi1.csv')
delhi2.to_csv('delhi2.csv')
delhi3.to_csv('delhi3.csv')
```
